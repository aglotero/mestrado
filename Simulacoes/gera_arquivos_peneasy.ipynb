```python
with open('tambor.geo', 'r') as f:
    geometria = f.read()
    
with open('tambor.in', 'r') as f:
    tambor = f.read()
```

```python
comandos = {}
```

```python
config = []
for n in n_histories:
    for nuclideo in nuclideos:
        for distancia in distancias_detetor:
            for energia in energia_inicial_multicanal:
                nome = "{0}_{1}cm_{2}estories_{3}".format(nuclideo.strip(),
                                                          distancia.strip(),
                                                          n.strip(),
                                                          energia.strip())
                
                print(nome)
                
                arquivo_input = tambor.format(n_histories=n,
                                              nuclideo=nuclideo,
                                              energia_inicial_multicanal=energia)
                
                arquivo_tambor = geometria.format(distancias_detetor=distancia)
                
                copy_tree('./tambor_original', "a_processar_v2/" + nome.replace('.', '-'))
                
                with open('./a_processar_v2/' + nome.replace('.', '-') + '/tambor.geo', 'w') as f:
                    f.write(arquivo_tambor)

                with open('./a_processar_v2/' + nome.replace('.', '-') + '/' + nome.replace('.', '-')+'.in', 'w') as f:
                    f.write(arquivo_input)
                    
                comandos[nome.replace('.', '-')] = 'source /home/andre/env.sh\ncd /export/ipen/PGF5356/novo_espectros/a_processar_v2/{nome}\nnohup peneasy_intel < {nome}.in\ncp tallyPulseHeightSpectrum.dat {destino}'.format(
                    nome=nome.replace('.', '-'),
                    destino='/export/ipen/PGF5356/novo_espectros/dados_v2/'+nome.replace('.', '-')+'.dat'
                )
```
